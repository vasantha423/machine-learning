```python
import pandas as pd

crop_data = pd.read_csv('/content/drive/MyDrive/FLASK ML/Crop/Crop_recommendation.csv')

crop_data
```

```python
def take_input():
    nitrogen = int(input("Enter nitrogen content: "))
    phosphorus = int(input("Enter phosphorus content: "))
    potassium = int(input("Enter potassium content: "))
    temperature = float(input("Enter temperature: "))
    humidity = float(input("Enter humidity: "))
    ph = float(input("Enter pH value: "))
    rainfall = float(input("Enter rainfall: "))

    return [nitrogen, phosphorus, potassium, temperature, humidity, ph, rainfall]
```

```python
def recommend_crop(input_values, crop_data):
    # Calculate the distance between the input values and each crop in the dataset
    distances = ((crop_data.iloc[:, :-1] - input_values)**2).sum(axis=1)
    
    # Find the index of the crop with the minimum distance
    min_index = distances.idxmin()
    
    # Return the label of the crop with the minimum distance
    recommended_crop = crop_data.loc[min_index, 'label']
    
    return recommended_crop
```

```python
# Take input from the user
user_input = take_input()

# Recommend a crop based on the user input
recommended_crop = recommend_crop(user_input, crop_data)

print("Recommended crop : ", recommended_crop)
```
